```python
import os
from google.colab import userdata

os.environ["GOOGLE_API_KEY"] = userdata.get("GEMINI_API_KEY")
```

```python
from google import genai

# The client gets the API key from the environment variable `GEMINI_API_KEY`.
client = genai.Client()

def ask_gemini(prompt):
    response = client.models.generate_content(model="gemini-3-flash-preview", contents = prompt)
    return response.text.strip()

def run_pipeline(text):
    # Step 1: Summarize
    summary = ask_gemini(f"""
    Summarize the following text in 3-5 bullet points:

    {text}
    """)

    # Step 2: Critique
    critique = ask_gemini(f"""
    Critique this summary. What is missing or unclear?

    {summary}
    """)

    # Step 3: Improve
    improved = ask_gemini(f"""
    Rewrite the summary incorporating the critique. Make it clearer and more complete.

    Summary:
    {summary}

    Critique:
    {critique}
    """)

    return summary, critique, improved
```

```python
text = """
Artificial intelligence is transforming industries by automating tasks,
improving decision-making, and enabling new capabilities. However, it also
raises ethical concerns such as bias, privacy, and job displacement.
"""

summary, critique, improved = run_pipeline(text)

print("SUMMARY:\n", summary)
print("\nCRITIQUE:\n", critique)
print("\nIMPROVED:\n", improved)
```
